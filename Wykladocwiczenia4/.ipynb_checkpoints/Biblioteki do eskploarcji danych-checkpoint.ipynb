    Statystyką stosowaną

Uczenie maszynowe polega na rozpoznawaniu wzorców. Schemat działania opiera się na 3 krokach:
    
    Wzorzec -> Przewidywanie -> Nauka
    
Posiadamy pewien wzorzec - poznajemy jego wynik i uczymy się związku pomiędzy tymi dwiema informacjami

Uczenie maszynowe można subiektywnie określić jako:

    Metoda by znaleźć związek między X i Y - funkcja mapująca
    
    Metoda przewidywania przyszłości - funkcja przewidująca
    
    Metoda rozpoznawania - funkcja klasyfikująca

```python
plt.figure(figsize=(8,8))
plt.scatter(x=blob[0][:,0],y=blob[0][:,1])
```

```python
algorytm_klasteryzacyjny = KMeans(n_clusters = 3)
```

```python
algorytm_klasteryzacyjny
```

```python
algorytm_klasteryzacyjny.fit(blob[0])
```

```python
etykiety = algorytm_klasteryzacyjny.labels_
print(etykiety)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x = blob[0][:,0], y = blob[0][:,1], c = etykiety)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x = blob[0][:,0], y = blob[0][:,1], c = etykiety, cmap = 'viridis')
```
