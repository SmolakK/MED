Eksploracja danych to proces zrozumienia danych i wyciągania wniosków. Istnieją różne podejścia, na przykład:

    KDD
    CRISP-DM

Proces knowledge discovery in databases (KDD) to ogólnie przyjęty schemat działań podczas pracy z danymi. Składają się na niego etapy:

    Selekcja
    Pre-processing
    Transformacja
    Wydobywanie danych
    Interpretacja/Ewaluacja

Inne podejście to Cross Industry Standard Process for Data Mining (CRISP-DM), który definiuje schemat pracy z danymi:

    Zrozumienie biznesowe
    Zrozumienie danych
    Przygotowanie danych
    Modelowanie
    Ewaluacja
    Wdrożenie

Wszystko sprowadza się do tego samego punktu. Ogólnie zatem proces eksploracji danych ująć można następująco:

    Zrozumienie danych 
    Przygotowanie danych
    Modelowanie
    Ewaluacja
    Wnioski/Interpretacja i ewentualne powtórzenie procesu eksploracji
