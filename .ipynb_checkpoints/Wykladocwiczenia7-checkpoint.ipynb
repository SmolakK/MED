## Brakujące wartości/ błędne pomiary

```python
import pandas as pd
import numpy as np
```

Dlaczego nie ma danych?

Są 3 powody dlaczego danych nie ma w zbiorze:

    Przypadek losowy - brak danych nie jest związany z samymi danymi, a jednynie właściwościami ich gromadzenia. Przykładowo: błąd urządzenia gromadzącego dane (może skutkować błędną wartością pomiaru)
    
    Przypadek całkowicie losowy - brak danych wynika jedynie z przypadku losowego - skasowany przypadkowo ze zbioru (wartość pusta lub błędnie wprowadzona)
    
    Przypadek nielosowy - celowo usunięte dane ze zbioru (np. ochrona danych)

Jak radzić sobie z brakującymi danymi? Są dwie metody:

    Usunąć
        
        LUB
    
    Obliczyć

Serie czasowe mogą składać się z trzech elementów:
    
    1. Trendu
    2. Sezonowości
    3. Rezyduów
    

```python
plt.plot(np.arange(temp_missing.iloc[40:80].shape[0]),temp_missing["Temp"].iloc[40:80])
```

```python
#Last Observation Carried Forward(LOCF)
#Next Observation Carried Backward(NOCB)
#Interpolacja
#Dostosowane sezonowe i interpolacja
```

```python
temp_filled = temp_missing.fillna(method="bfill")
plt.plot(np.arange(temp_filled.iloc[40:80].shape[0]),temp_filled["Temp"].iloc[40:80])
```

```python
temp_filled = temp_missing.interpolate("linear")

plt.plot(np.arange(temp_filled.iloc[40:80].shape[0]),temp_filled["Temp"].iloc[40:80])
```

```python
temp_fill = temp_missing.copy()
temp_fill = temp_fill.fillna(temp_fill.shift(12))
plt.plot(np.arange(temp_fill.iloc[40:80].shape[0]),temp_fill["Temp"].iloc[40:80])
```
